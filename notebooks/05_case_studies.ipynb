# 05. Case Studies

This notebook produces:
- **Graph 8**: Detailed case study plots for top 3 spike events

Each case study shows:
- Price behavior 6 hours before and after
- Demand changes
- Renewable generation (if available)
- Context and explanation

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from src import config
from src import io as data_io
from src import features
from src import plots

%matplotlib inline
plt.style.use('seaborn-v0_8-whitegrid')
```

## 1. Load Data

```python
# Load main data
try:
    df = data_io.load_processed_data('nsw_with_volatility')
except FileNotFoundError:
    df = data_io.load_processed_data('nsw_processed')

df = features.compute_time_features(df)

# Add renewable share if possible
if 'solar' in df.columns or 'wind' in df.columns:
    df = features.compute_renewable_share(df)

print(f"Data: {len(df)} intervals")
print(f"Columns: {list(df.columns)}")
```

```python
# Load or recreate event table
try:
    event_df = pd.read_csv(config.DATA_PROCESSED / 'spike_events.csv', parse_dates=['start_time', 'end_time'])
    print(f"Loaded {len(event_df)} spike events")
except FileNotFoundError:
    print("Event table not found. Recreating...")
    is_spike = features.detect_spikes_absolute(df)
    event_ids = features.merge_spike_events(is_spike)
    event_df = features.create_spike_event_table(df, event_ids)
    event_df = features.compute_pre_spike_features(df, event_df)
```

```python
# View event summary
print("Event statistics:")
display(event_df.describe())

print("\nTop 10 events by max price:")
display(event_df.nlargest(10, 'max_price')[['start_time', 'end_time', 'duration_min', 'max_price', 'mean_price']])
```

## 2. Select Case Studies

Criteria for selection:
1. **Top 3 by max price** - most extreme events
2. Alternatively: diverse selection across months/causes

```python
# Select top 3 events by max price
top_events = event_df.nlargest(3, 'max_price')

print("Selected case study events:")
for i, (_, event) in enumerate(top_events.iterrows(), 1):
    print(f"\nCase {i}:")
    print(f"  Date: {event['start_time'].strftime('%Y-%m-%d %H:%M')}")
    print(f"  Duration: {event['duration_min']} minutes")
    print(f"  Max Price: ${event['max_price']:.2f}/MWh")
    print(f"  Mean Price: ${event['mean_price']:.2f}/MWh")
```

## 3. Case Study 1

```python
event1 = top_events.iloc[0]

print(f"CASE STUDY 1: {event1['start_time'].strftime('%Y-%m-%d %H:%M')}")
print(f"="*60)
print(f"Max Price: ${event1['max_price']:.2f}/MWh")
print(f"Duration: {event1['duration_min']} minutes")

# Plot
fig = plots.plot_case_study(df, event1, hours_before=6, hours_after=6, case_number=1)
plt.show()
```

```python
# Detailed analysis of Case 1
start = event1['start_time'] - pd.Timedelta(hours=1)
end = event1['end_time'] + pd.Timedelta(hours=1)
case1_data = df.loc[start:end]

print("\nPre-spike conditions (1 hour before):")
pre_spike = df.loc[event1['start_time'] - pd.Timedelta(hours=1):event1['start_time']]
print(f"  Price: ${pre_spike['rrp'].iloc[0]:.2f} -> ${pre_spike['rrp'].iloc[-1]:.2f}")
if 'total_demand' in pre_spike.columns:
    print(f"  Demand: {pre_spike['total_demand'].iloc[0]:.0f} -> {pre_spike['total_demand'].iloc[-1]:.0f} MW")
if 'solar' in pre_spike.columns:
    print(f"  Solar: {pre_spike['solar'].iloc[0]:.0f} -> {pre_spike['solar'].iloc[-1]:.0f} MW")
if 'wind' in pre_spike.columns:
    print(f"  Wind: {pre_spike['wind'].iloc[0]:.0f} -> {pre_spike['wind'].iloc[-1]:.0f} MW")
```
